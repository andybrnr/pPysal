```python
figsize(8,8)
plot(ds, queen_bin,'ro-', label='bin')
plot(ds, queen_rt,'go-', label='rtree')
plot(ds, queen_lis,'yo-', label='list')
xlabel('Polygon Count')
ylabel('Time (sec.)')
grid()
legend(loc=2)
```
